```python
import h2o
```

```python
from h2o.automl import H2OAutoML
# H2o autoML can do data pre-processing: take care of missing values, categorical encoding conversion
# It also provides a deployment code /
```

```python
h2o.init()
#Initiates de H2o cluster
```

```python
# Load data
economic_data = '/Users/PlumpsCrap/Downloads/data_LK_quarterly.csv'
df = h2o.import_file(economic_data)
```

```python
df.types
```

```python
df.describe()
```

```python
# Dependent variable 
y = "CAN GDP QOQ"
```

```python
# Split data
splits = df.split_frame(ratios=[0.7],seed=1)
train = splits[0]
test = splits[1]
```

```python
# you can run the autoML limiting the max # of models (max_models=10) or ...
#excluding some other models (exclude_algos=["StackedEnsemble","DeepLearning"])
# StackedEnsamble model is one of the best in H2o, but its not simple and could complicate 
# Verosity="info" is to get an output that you can debug 

aml = H2OAutoML(max_runtime_secs=300,seed=1,project_name="DS_Project_ForecastEconomy",verbosity="info",exclude_algos=["StackedEnsemble","DeepLearning"])
aml.train(y = y, training_frame=train, leaderboard_frame= test)
```

```python
aml.leaderboard.head()
```

```python
#Predictions
pred = aml.predict(test)
pred.head()
```

```python
#Generete a performance report of the best model

perf = aml.leader.model_performance(test)
perf
```

```python
# List all the model IDs to have the option analyze all the models 

model_ids = list(aml.leaderboard['model_id'].as_data_frame().iloc[:,0])
```

```python
model_ids
```

```python
#Analyze model Stacked
out = h2o.get_model([mid for mid in model_ids if 'Stacked' in mid][0])
```
